## Matplotlib 연습

### Matplotlib, Seaborn 활용

```python
# 필요 라이브러리 import
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# 한글 꺠짐문제 해결
from matplotlib import rc

rc('font', family='D2Coding')

plt.rcParams['axes.unicode_minus'] = False
```

### 타이타닉 데이터

```python
sns.get_dataset_names()
```

```python
titanics = sns.load_dataset('titanic')
```

```python
titanics.info()
```

```python
titanics.describe() # 숫자만 있는 필드만 나온다.
```

```python
pclassSurvived = pd.crosstab(titanics['pclass'], titanics['survived'])
pclassSurvived
```

```python
# 컬럼명을 매핑
pclassSurvived.columns = pclassSurvived.columns.map({0:'Dead', 1:'Alive'})
```

```python
pclassSurvived
```

```python
sns.set_style('darkgrid')
```

```python
# bar 차트
pclassSurvived.plot.bar() 
pclassSurvived.plot.bar(subplots=True)

plt.show()
```

```python
# 가로바 차트
pclassSurvived.plot.barh()
plt.show()
```

```python
# 라인차트
pclassSurvived.plot.line()
plt.show()
```

```python
# 파이차트
pclassSurvived['Dead'].plot.pie(title='등급객실 별 사망자비율')

plt.show()
```

```python
sns.barplot(pclassSurvived)
plt.show()
```
