### Mac 에서 한글 깨짐 현상 해결

```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import rc

rc('font', family='D2Coding')

plt.rcParams['axes.unicode_minus'] = False
```

```python
# 단일차트
plt.plot([1, 2, 3, 4, 5])
plt.show()
```

```python
plt.plot(np.random.rand(50))  # 50개의 정수를 랜덤하게 만든다.
plt.show()
```

```python
# 다중차트, figure() 사용
fig = plt.figure()
plt1 = fig.add_subplot(2, 2, 1)  # 1번 차트
plt2 = fig.add_subplot(2, 2, 2)  # 2번 차트
plt3 = fig.add_subplot(2, 2, 3)  # 3번 차트
plt4 = fig.add_subplot(2, 2, 4)  # 4번 차트

plt1.plot(np.random.randn(50))
plt2.hist(np.random.randn(100))  # 히스토그램
plt3.scatter(np.arange(30), np.random.randn(30))  # scatter차트 (x축, y축)
plt4.plot(np.random.randn(30), 'violet')  # 색상 추가 가능

plt.show()
```

```python
# 다중차트 - 2번째
# figsize 옵션으로 넓이,높이 조정가#
fig, axes = plt.subplots(2, 2, figsize=(14, 6))  # row, column
axes[0][0].plot(np.random.randn(50))
axes[0][1].hist(np.random.randn(100))
axes[1][0].scatter(np.arange(30), np.random.randn(30))
axes[1][1].plot(np.random.randn(30), 'violet')

plt.show()
```

### 차트 스타일 지정 옵션
- `label` , `style` , `alpha` , `rot` , `xticks` , `yticks` , `xlim` , `ylim` , `color` , `marker` ...

```python
fig, axes = plt.subplots(2, 3, figsize=(14, 6))
# color = 'k' 는 검정색 , alpha 는 투명도 지정(0.0 ~ 1.0)
axes[0][0].plot(np.random.randn(50), color='c', alpha=0.6)

# bins 는 grouping 갯수
axes[0][1].hist(np.random.randn(100), color='green', bins=10)

# linestyle 은 선 종류 : -- , --, -. :, solid , dashed , dashdot , dotted
axes[0][2].plot(np.random.randn(50), color='r', linestyle='dotted')

# marker : *(별) / ^(삼각형) / o(동그라미)
axes[1][0].scatter(np.arange(30), np.random.randn(30), marker='*')
axes[1][1].plot(np.random.randn(30), 'violet')
axes[1][2].plot(np.random.randn(20), marker="o")

plt.show()
```

```python
# cumsum() : cumulative sum 누적합
plt.plot(np.random.randn(50).cumsum(), label='data1')
plt.plot(np.random.randn(50).cumsum(), label='data2')
plt.plot(np.random.randn(50).cumsum(), label='data3')

# legend() : 범례 표시함수
#  'best', 'upper right', 'upper left', 'lower left', 'lower right', 'right' 
#  'center left', 'center right', 'lower center', 'upper center', 'center'
plt.legend(loc='lower left')
plt.show()
```

```python
# x,y축 설정
plt.figure(figsize=(18, 5))
plt.plot(np.random.randn(1000))

# x 축을 0, 250, 500, 750, 1000 으로 나눈다.
plt.xticks([0, 250, 500, 750, 1000], labels=['min','25%','median','75%','max']) 

plt.title('Stastics') # 한글에는 문제가 있다.
plt.show()
```
